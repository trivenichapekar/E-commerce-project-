## Problem Statement

### A project with an Ecommerce company sells clothing online but they also have in-store style and clothing advice sessions. Customers come in to the store, have sessions/meetings with a personal stylist, then they can go home and order either on a mobile app or website for the clothes they want. **The company is trying to decide whether to focus their efforts on their mobile app experience or their website.** They have asked to help them figure it out.


•	Avg. Session Length: Average session of in-store style advice sessions in minutes.

•	Time on App: Average time spent on App in minutes

•	Time on Website: Average time spent on Website in minutes

•	Length of Membership: How many years the customer has been a member.

•	Yearly Amount Spent: The total amount the customer is spending in dollars.

## Model Building

```python
X = df1.iloc[:,:-1]
Y= df1.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
test=y_test
```

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the X_train and X_test data
X_train = scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)

# Fit and transform the y_train and y_test data
y_train = scaler.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_test = scaler.transform(y_test.values.reshape(-1, 1)).flatten()
```

```python
#Linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt
```

```python
# Initialize the Linear Regression model
modell = LinearRegression()
# Fit the model on the training data
modell.fit(X_train, y_train)
# Predict the target variable on the test set
y_pred = modell.predict(X_test)
# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
# Calculate RMSE
rmse = sqrt(mean_squared_error(y_test, y_pred))
# Print the results
print("R-squared Score:", r2)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
#Polynomial regression
from sklearn.preprocessing import PolynomialFeatures
# Apply polynomial features to the input data
degree = 2  # Degree of the polynomial
poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)
# Initialize the Linear Regression model
modelp = LinearRegression()
# Fit the model on the training data with polynomial features
modelp.fit(X_train_poly, y_train)
# Predict the target variable on the test set with polynomial features
y_pred = modelp.predict(X_test_poly)
# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
# Calculate RMSE
rmse = sqrt(mean_squared_error(y_test, y_pred))
# Print the results
print("R-squared Score:", r2)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
#lasso Regression
from sklearn.linear_model import Lasso
# Initialize the Lasso Regression model
alpha = 1.0  # Regularization strength (adjust as needed)
modella = Lasso(alpha=alpha)
# Fit the model on the training data
modella.fit(X_train, y_train)
# Predict the target variable on the test set
y_pred = modella.predict(X_test)
# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
# Calculate RMSE
rmse = sqrt(mean_squared_error(y_test, y_pred))
# Print the results
print("R-squared Score:", r2)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
# Initialize the Ridge Regression model
alpha = 0.5  # Regularization strength (adjust as needed)
modelr = Ridge(alpha=alpha)
# Fit the model on the training data
modelr.fit(X_train, y_train)
# Predict the target variable on the test set
y_pred = modelr.predict(X_test)
# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
# Calculate RMSE
rmse = sqrt(mean_squared_error(y_test, y_pred))
# Print the results
print("R-squared Score:", r2)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
#Elastic net Regression
from sklearn.linear_model import ElasticNet
# Initialize the Elastic Net Regression model
alpha = 1.0  # Regularization strength (adjust as needed)
l1_ratio = 0.5  # L1 regularization ratio (adjust as needed)
modele = ElasticNet(alpha=alpha, l1_ratio=l1_ratio)
# Fit the model on the training data
modele.fit(X_train, y_train)
# Predict the target variable on the test set
y_pred = modele.predict(X_test)
# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
# Calculate RMSE
rmse = sqrt(mean_squared_error(y_test, y_pred))
# Print the results
print("R-squared Score:", r2)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
#Decision tree regression
from sklearn.tree import DecisionTreeRegressor
# Initialize the Decision Tree Regression model
modeld = DecisionTreeRegressor()
# Fit the model on the training data
modeld.fit(X_train, y_train)
# Predict the target variable on the test set
y_pred = modeld.predict(X_test)
# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
# Calculate RMSE
rmse = sqrt(mean_squared_error(y_test, y_pred))
# Print the results
print("R-squared Score:", r2)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
# Random forest regression
from sklearn.ensemble import RandomForestRegressor
# Initialize the Random Forest Regression model
modelrf = RandomForestRegressor(n_estimators=100, random_state=42)
# Fit the model on the training data
modelrf.fit(X_train, y_train)
# Predict the target variable on the test set
y_pred = modelrf.predict(X_test)
# Evaluate the model accuracy
r_squared = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = sqrt(mse)
# Print the accuracy metrics
print("R-squared:", r_squared)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
# Support vector regressor
from sklearn.svm import SVR
# Initialize the Support Vector Regression model
modelsv = SVR(kernel='rbf')  # You can choose different kernels (linear, radial basis function, polynomial, etc.)
# Fit the model on the scaled training data
modelsv.fit(X_train, y_train)
# Predict the target variable on the scaled test set
y_pred = modelsv.predict(X_test)
# Evaluate the model accuracy
r_squared = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = sqrt(mse)
# Print the accuracy metrics
print("R-squared:", r_squared)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
#KNN
from sklearn.neighbors import KNeighborsRegressor
# Initialize the KNN Regression model
knn_model = KNeighborsRegressor(n_neighbors=4)  # You can choose different values for k
# Fit the model on the scaled training data
knn_model.fit(X_train, y_train)
# Predict the target variable on the scaled test set
y_pred = knn_model.predict(X_test)
# Evaluate the model accuracy
r_squared = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = sqrt(mse)
# Print the accuracy metrics
print("R-squared:", r_squared)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor
# Initialize the base estimator (Random Forest)
base_model = RandomForestRegressor(n_estimators=150, random_state=42)
# Initialize the Bagging Regressor
bagging_model = BaggingRegressor(base_model, n_estimators=50, random_state=42)
# Fit the model on the scaled training data
bagging_model.fit(X_train, y_train)
# Predict the target variable on the scaled test set
y_pred = bagging_model.predict(X_test)
# Evaluate the model accuracy
r_squared = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = sqrt(mse)
# Print the accuracy metrics
print("R-squared:", r_squared)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
from sklearn.ensemble import AdaBoostRegressor
# Initialize the base estimator (Ridge Regression)
base_model = DecisionTreeRegressor(max_depth=3)
# Initialize the AdaBoost Regressor with Ridge as the base model
adaboost_model = AdaBoostRegressor(base_model, n_estimators=50, random_state=42)
# Fit the model on the scaled training data
adaboost_model.fit(X_train, y_train)
# Predict the target variable on the scaled test set
y_pred_adaboost = adaboost_model.predict(X_test)
# Evaluate the AdaBoost model accuracy
r_squared_adaboost = r2_score(y_test, y_pred_adaboost)
mse_adaboost = mean_squared_error(y_test, y_pred_adaboost)
rmse_adaboost = sqrt(mse_adaboost)
# Print the AdaBoost model accuracy metrics

print("AdaBoost Model with Decision Tree model - R-squared:", r_squared_adaboost)
print("AdaBoost Model with Decision Tree model - Mean Squared Error (MSE):", mse_adaboost)
print("AdaBoost Model with Decision Tree model- Root Mean Squared Error (RMSE):", rmse_adaboost)
```
